```python
# IPL 2025 Data Analysis (Python Project)

### Author: DAYAKAR  
### Tools: Python, Pandas, Seaborn, Matplotlib

## Objective:
#To analyze batting and bowling performance in IPL 2025 and extract key insights.

##--------------------------------------------------------------------------------------------#
# Import required libraries
import pandas as pd

batters = pd.read_csv("C:/Users/J Dayakar/Downloads/ipl 2025 data set/IPL2025Batters.csv")
bowlers = pd.read_csv("C:/Users/J Dayakar/Downloads/ipl 2025 data set/IPL2025Bowlers.csv")

print(batters.head())
print(bowlers.head())
```
